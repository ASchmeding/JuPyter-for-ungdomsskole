## Oppgaver med løsning som kan vises

Noen ganger kan det være ønskelig at man legger ved løsninger til oppgaver. På den måten kan elever få hjelp hvis de ikke klarer oppgaven eller hvis de vil se en annen løsning av oppgaven.
Vanligvis ønsker vi da å gi dem en mulighet å vise løsning men løsningen skulle ikke være synlig direkte fra begynnelsen.

#### Kommentarfunksjon
Det fins flere muligheter å gjemme løsninger slik at de er tilgjengelig etter man trykker på en knapp eller liknende.
En av de enkleste muligheter for dette er å bruke kommentarfunksjonen i Markdown. Vi viser det i det neste celle (bruk dobbeltklikk for å vise koden):

Dobbeltklikk __her__ for løsning.

<!-- Din løsnign står under:

Tekst av løsning eller kommentar. Her bruker vi faktisk Markdowns kommentar funksjon. Dobbeltklikk aktiverer editering av tekstcellen og gjør kommentar synlig. Om vi ønsker å gå tilbake til opprinnelig visnign må vi bruke SHIFT+ENTER mens cellen er aktivert. 

-->


Å bruke Markdowns kommentarfunksjon er en av de enkleste muligheter, men det har ulempen at vi bruker egentlig en kommentarfunksjon og får tilbake uformatert Markdown (som vi kanskje ikke ville vise til elever). 

#### `<details>` kommando
Markdown har også noen innebyggde funksjoner som kan hjelpe med tekst som skulle være synlig etter brukeren trykker på en knapp (eller liknende).
For eksempel kan vi bruke `<details>` kommando i Markdown. Den vises nede:
````md
<details>
  <summary>Trykk her for å vise</summary>
  
  ### Hva som helst
  1. Tekst!
  2. Andre ting som du ville vise
</details>

````
Resultatet ser sån ut:

<details>
  <summary>Trykk her for å vise</summary>

  ### Hva som helst
  1. Tekst!
  2. Andre ting som du ville vise
</details>
