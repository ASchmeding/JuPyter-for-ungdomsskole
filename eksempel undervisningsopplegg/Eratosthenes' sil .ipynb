### Eratosthenes' sil

**Problem** Hvordan kan vi finne en liste av alle primtall som er mindre eller lik et gitt tall $n$.

For eksempel hvis $n=10$, kan vi lett sjekke om alle tall opp til $10$ er primtall. 
Siden primtall er større enn $1$ starter vi med $2$ som er prim. Også $3$ er prim, men siden $4 = 2 \cdot 2$ er $4$ ikke prim. Vi kan jobbe oss videre og finner ut at de neste to primtallene er $5,7$, dermed har vi en fullstendig liste av alle primtall mindre enn $10$ som ser slik ut
$$2,3,5,7$$

Eratosthenes, en gresk matematikker, beskrev en algoritme som viser hvordan man kan finne alle primtall mindre enn $n$ på en systematisk måte. Algoritmen er nå kjent som Eratosthenes' sil og er gitt i ord som følgende

#### Algoritmen

For å finne alle primtall mindre enn eller lik et gitt heltall $n$ ved hjelp av Eratosthenes' metode:

1. Lag en liste av påfølgende heltall fra $2$ til $n$: $(2, 3, 4, \ldots, n)$.
2. La $p$ være lik til $2$, det første primtallet, i begynnelsen.
3. Stryk ut alle multipler av $p$ som er større enn eller lik $p\cdot p$, fra listen.
4. Finn det første tallet større enn $p$ som står igjen på listen. Dette tallet er det neste primtallet. Sett $p$ lik dette tallet.
5. Gjenta trinn 3 og 4 inntil $p$ er større enn $n$.

Alle gjenværende tall på listen er primtall.
