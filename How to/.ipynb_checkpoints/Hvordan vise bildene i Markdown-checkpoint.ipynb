# Bildene i JuPyter og Markdown

I denne notat skal vi se nærmere om hvordan man kan legge inn bildene i JuPyter notater.
Det fins flere muligheter og vi skal vise dem frem i de neste celler. Vi anbefaler å dobbelklikke på bildene om du lurer å se kode som ble brukt (det er enklest å stjele koden og tilpasse det bare til det du ønsker å gjøre.

## Metode 1: Legg inn bildene i JuPyter notat

Det enkleste metoden er å legge inn en ny Markdown celle for bildet du ønsker å legge inn. Deretter kan du bruke "Insert Image" fra JuPyter meny i fanen "Edit" som gir mulighet å velge et bilde fra datamaskinen din. 

Enda enklere kan du bare bruke "Drag and Drop", dvs. dra bildet fra mappen den ligger i inn Markdown cellen. JuPyter gjør deretter resten av arbeidet. Resultatet kan du se i neste cellen:

Dobbeltklikk på bildet for å se kode. Hvis du bruker for eksempel Google Colab for å vise denne JuPyter notat ville du ikke se bildet over (fordi Google ikke støtter visning av vedlagte bildene). Det er også den største ulempe metoden har.

Om du ønsker å legge inn et bilde som skulle være synlig også ved bruk av for eksempel Google Colab bruk en av de metoder som følger nå.

## Metode 2: Bruk lenke til bildet på nett

Det er enkelt å bruke lenker i Markdown for å vise bildene i JuPyter notater.
Grunnleggende syntaks for det er

`![alternativ tekst](URL-som-peker-på-bilde)`

Det kan gjøres for eksempel med bildene som ligger på Wikipedia. Som illustrasjon la oss anta at vi ønsker å legge inn et bilde til den berømte norsk matematiker Idun Reiten (bildet fins på Wikipedia.no og er tilgjenglig ved CC BY-SA 2.0 DE lisens). Resultaten til lenking ser du i neste cellen.

![Idun Reiten](https://upload.wikimedia.org/wikipedia/commons/4/45/Idun_Reiten.jpg)

Alternativ kan man også bruke HTML syntaks for disse lenker:

`<img src="URL-som-peker-på-bildet" alt="Alternativ tekst" />`

Mer om HTML syntaks for lenker og muligheter man har for innstillinger (for eksempel kan man endre størrelse og liknende for bildene) fins her https://www.w3schools.com/html/html_images.asp

### Lenker til bildene på GitHub

Bildene som ligger på GitHub er litt spesiell. Om vi bruker for eksempel HTML syntaks vist oppe,

`<img src="https://github.com/ASchmeding/Intro-2-Python-norsk/blob/main/Images/TuplesQuiz.png" alt="Noen bildet fra GitHub" />`

får vi dette resultat:

<img src="https://github.com/ASchmeding/Intro-2-Python-norsk/blob/main/Images/TuplesQuiz.png" alt="Noen bildet fra GitHub" />

Det samme skjer når vi bruker alternativ syntaks `![Noen bildet fra GitHub](https://github.com/ASchmeding/Intro-2-Python-norsk/blob/main/Images/TuplesQuiz.png)` gir oss:

![Noen bildet fra GitHub](https://github.com/ASchmeding/Intro-2-Python-norsk/blob/main/Images/TuplesQuiz.png)

Dessverre vises det ikke bildet men bare beskrivelsen vi har gitt. Grunnet til det er hvordan GitHub handterer bildene som ligger i GitHub mapper. Det fins et enkelt mulighet å vise bildene som ligger på GitHub uten at vi må endre mye. For dette må vi bare endre lenken i kode litt:

`<img src="URL-som-peker-på-bildet?raw=true" alt="Alternativ tekst" />`

Som vist har vi brukt samme lenke men har lagt til "?raw=true" til endelse av lenken. Om vi gjør det er resultatet det følgende:

<img src="https://github.com/ASchmeding/Intro-2-Python-norsk/blob/main/Images/TuplesQuiz.png?raw=true" alt="Noen bildet fra GitHub" />
